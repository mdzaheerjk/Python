```python
class OurBankAccount:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance

    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficient Balance")
        else:
            self.balance-=amount
            print(self.name,"withdraw",amount,"new balance",self.balance)
```

```python
user_1=OurBankAccount("zaheer",10000)
```

```python
user_1.withdraw(100)
```

```python
print(type(user_1))
```

```python
user_2=OurBankAccount("Owais",1000)
```

```python
user_2.withdraw(500)
```

```python
user_2.withdraw(600)
```

```python
class Robot:
    def __init__(self,customer_name=None):
        print("Hello Welcome to Robot!")
        self.name="Chitti"
        self.ram_in_gb=10
        self.version="0.0.1"
        self.customer_name=customer_name

    def config(self):
        print("Ram :",self.ram_in_gb)
        print("Version :",self.version)
        print("Robot name :",self.name)
    def customer_details(self):
        print("customer name is :",self.customer_name)
```

```python
r1=Robot("zaheer")
```

```python
r1.config()
```

```python
r1.customer_details()
```

```python
r2=Robot("Nikhil")
```

```python
r2.config()
```

#### Encapsulation

- Public : {"syntax" : "name", "Access" : "anywhere"}
- Protected : {"syntax" : "_name", "Access" : "class + child classes"}
- Private : {"syntax" : "__name", "Access" : "only inside class"}

public, protected and private assumes the developer is responsible. Consent.

```python
class BankAccount:
    def __init__(self,name,account_type,balance):
        self.name=name
        self._account_type=account_type
        self.__balance=balance
    def get_balance(self):
        print("Current balance is :",self.__balance)
    def add_and_get_balance(self,amount):
        self.__balance+=amount
        print("New Balance is :",self.__balance)
```
