# control flow

```python
names=["Sudarshan","Praveen","Abhijeet","zaheer","Sandeep"]
```

```python
for name in names:
    if name=='zaheer':
        continue
    name=name.capitalize()
    name=name.upper()
    print(name)
```

```python
for name in names:
    if name=="zaheer":
        break
    name=name.capitalize()
    name=name.upper()
    print(name)
```

```python
count=0
print("Start")
while True:
    if count>100:
        break
    print(count)
    count+=1
print('End')
```

```python
for i in range(1,11):
    print(i)
    for name in names:
        if name in ["Praveen","Abhijeet","zaheer","Sandeep"]:
            break
        print(name)
```

```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def multi(a,b):
    print(a*b)
def divide(a,b):
    print(a/b)

def calculator():
    choice=int(input("enter 1 for addition , 2 for substraction , 3 for multiplication and 4 for division"))
    value1=int(input("enter the first number"))
    value2=int(input("enter the second number"))
    
    if choice==1:
        add(value1,value2)
    elif choice==2:
        sub(value1,value2)
    elif choice==3:
        multi(value1,value2)
    elif choice==4:
        divide(value1,value2)
    else:
        print("incorrect choice")

```

```python
calculator()
```
