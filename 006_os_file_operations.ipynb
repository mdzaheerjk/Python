```python
def add(a=0,b=0):
    c=a+b 
    return c 
```

```python
def add(*a):
    print(type(a))
    print(a)
    sum=0
    for i in a:
        sum+=i 
    print(sum)
```

```python
add(1,2,3,4,5,6,7,8,9)
```

```python
add(1)
```

```python
add(1,123.0,100,10023242,345)
```

```python
def add_v2(a,b,*more_values):
    print(type(a))
    print(a)
    print("="*10)
    print(type(b))
    print(b)
    print("="*10)
    print(type(more_values))
    print(more_values)
```

```python
add_v2(1,2,3,4,5)
```

```python
def add_v3(a,b=0,*args,**kwargs):
    print(type(a))
    print(a)
    print("="*10)
    print(type(b))
    print(b)
    print("="*10)
    print(type(args))
    print(args)
    print("*"*10)
    print(type(kwargs))
    print(kwargs)
```

```python
add_v3(1,2,1,2,3,4,5,6,7,8,9,0,value1=100,value=200)
```

```python
def order_pizza(size=5,*toppings,**details):
    print(size)
    print(toppings)
    print(details)
```

```python
def our_join_function(operating_system,*args):
    file_path=""
    for i in args:
        if operating_system=="windows":
            file_path=file_path+i+"\\"
        elif operating_system=="macos":
            file_path=file_path+i+'/'
    file_path=file_path[:-1]
    print(file_path)
    
```

```python
our_join_function("windows",os.getcwd(),"temp","dummy.txt")
```
